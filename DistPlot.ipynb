```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
a = sns.load_dataset("penguins")
```

```python
a
```

```python
sns.distplot(a.body_mass_g)
plt.show()
```

```python
sns.distplot(a.body_mass_g, rug = True)
plt.show()
```

```python
sns.distplot(a.body_mass_g, kde = False, rug = False, norm_hist=True)
plt.show()
```

```python
sns.distplot(a.body_mass_g, kde = True, rug = True)
plt.show()
```

```python
sns.distplot(a.body_mass_g, kde = True, rug = True, kde_kws = {'color':'Red', 'lw':2, 'alpha':0.9},
             hist_kws = {'color':'Blue', 'alpha':0.9, 'histtype':'step'},
             rug_kws = {'color':'Blue', 'height':0.2})
plt.show()
```
